```python
# Task 1: Load and Explore the Dataset

from ucimlrepo import fetch_ucirepo

# fetch dataset
bank_marketing = fetch_ucirepo(id=222)

# Load the dataset into a pandas DataFrame
df = bank_marketing.data.original

# Display the first few rows
print(df.head())
```

```python
# Task 2: Basic Data Analysis

# Basic statistics for numerical columns
print("Basic Statistics:")
print(df.describe())

# Group by education and calculate mean balance
print("\nAverage Balance by Education Level:")
print(df.groupby('education')['balance'].mean())
```

```python
# Task 3: Data Visualization

import matplotlib.pyplot as plt
import seaborn as sns

# Set a style
sns.set(style="whitegrid")

# 1. Line Chart - Duration over Day
plt.figure(figsize=(8, 4))
sns.lineplot(x='day_of_week', y='duration', data=df, marker='o')
plt.title("Call Duration Over Days")
plt.xlabel("Day of the Month")
plt.ylabel("Call Duration (seconds)")
plt.tight_layout()
plt.show()

# 2. Bar Chart - Average Balance by Education Level
plt.figure(figsize=(8, 4))
sns.barplot(x='education', y='balance', data=df, estimator='mean', palette="muted")
plt.title("Average Balance by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Average Balance")
plt.tight_layout()
plt.show()

# 3. Histogram - Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], bins=5, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 4. Scatter Plot - Balance vs Duration
plt.figure(figsize=(8, 4))
sns.scatterplot(x='balance', y='duration', hue='y', data=df)
plt.title("Balance vs Duration")
plt.xlabel("Account Balance")
plt.ylabel("Call Duration")
plt.legend(title='Subscribed')
plt.tight_layout()
plt.show()
```
